# Improving the Aesthetics of Matplotlib Charts

In this guided project, we'll be visualizing the gender gap across college degrees. The line charts generated will be used to make comparison across all degree categories. In the last step, we'll explore how to export the final diagram we create as an image file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
women_degrees[:7]
```

That is a snippet of the dataset we will be exploring with visualization in subsequent cells.

## Visualization in three-column format..

Each column represents a discipline that is either stem or liberal arts. There are some that do not fall in these two groups. This is the basis on which I will categorize the analysis.

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

grid = plt.figure(figsize=(18, 20))
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

for i in range(0, 6):    #for stem_cat
    ax = grid.add_subplot(6, 3, 1+(3*i))
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[i]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[i]], c=cb_orange, linewidth=3)
    ax.tick_params(left=False, bottom=False)
    ax.set_title(stem_cats[i])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if i==0:
        ax.text(2006, 79, 'Women')
        ax.text(2008, 18, 'Men')
    if i==5:
        ax.text(2009, 90, 'Men')
        ax.text(2007, 7, 'Women')
        
for i in range(0, 5):    #lib_arts_cats
    ax = grid.add_subplot(6, 3, 2+(3*i))
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[i]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[i]], c=cb_orange, linewidth=3)
    ax.tick_params(left=False, bottom=False)
    ax.set_title(lib_arts_cats[i])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if i==0:
        ax.text(2006, 73, 'Women')
        ax.text(2009, 24, 'Men')
        
for i in range(0, 6):    #other_cats
    ax = grid.add_subplot(6, 3, 3+(3*i))
    ax.plot(women_degrees['Year'], women_degrees[other_cats[i]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[i]], c=cb_orange, linewidth=3)
    ax.tick_params(left=False, bottom=False)
    ax.set_title(other_cats[i])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if i==0:
        ax.text(2006, 87, 'Women')
        ax.text(2009, 10, 'Men')
    if i==5:
        ax.text(2009, 60, 'Men')
        ax.text(2009, 38, 'Women')
        
```

The chart representation of the dataset is obviously too clustered. In the next few cells, I will be modifying some parameters to make the real analysis stick out to the reader.

### Removing x-axis
In the next analysis, I will be removing the x axis of each chart except the last one in each column to decluster the image.

```python
grid = plt.figure(figsize=(18, 20))
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

for i in range(0, 6):    #for stem_cat
    ax = grid.add_subplot(6, 3, 1+(3*i))
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[i]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[i]], c=cb_orange, linewidth=3)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.set_title(stem_cats[i])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if i==0:
        ax.text(2006, 79, 'Women')
        ax.text(2008, 18, 'Men')
    if i==5:
        ax.text(2009, 90, 'Men')
        ax.text(2007, 7, 'Women')
ax.tick_params(labelbottom='on')
        
for i in range(0, 5):    #lib_arts_cats
    ax = grid.add_subplot(6, 3, 2+(3*i))
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[i]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[i]], c=cb_orange, linewidth=3)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.set_title(lib_arts_cats[i])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if i==0:
        ax.text(2006, 73, 'Women')
        ax.text(2009, 24, 'Men')
ax.tick_params(labelbottom='on')        

for i in range(0, 6):    #other_cats
    ax = grid.add_subplot(6, 3, 3+(3*i))
    ax.plot(women_degrees['Year'], women_degrees[other_cats[i]], c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[i]], c=cb_orange, linewidth=3)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.set_title(other_cats[i])
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    if i==0:
        ax.text(2006, 87, 'Women')
        ax.text(2009, 10, 'Men')
    if i==5:
        ax.text(2009, 60, 'Men')
        ax.text(2009, 38, 'Women')
ax.tick_params(labelbottom='on')    
```

### Declustering the y-axis
